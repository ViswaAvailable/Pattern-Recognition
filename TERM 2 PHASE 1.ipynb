```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#2 class linear seperable classification
X = np.array(np.linspace(-5, 5, 1000))
X = np.array([[x, 3*x-5+3*np.random.randn()] for x in X])
Y = X[:,1]>3*X[:,0]-5
Y = np.array([int(y) for y in Y])
cols = ['b', 'r', 'g', 'm', 'c']
c = [cols[int(i)] for i in Y]
plt.scatter(X[:,0], X[:,1], color=c)
plt.show()
```

```python
#activation function=linear
#optimizer=Adam
#loss function=MeanSquaredError
model_1 = keras.Sequential([keras.layers.Dense(1, input_shape = (2,))])
model_1.compile(optimizer = 'Adam', loss = tf.keras.losses.MeanSquaredError(), metrics = ['accuracy'])
hist = model_1.fit(X,Y, epochs = 1000, verbose = 0)
```

```python
loss = hist.history['loss']
acc = hist.history['accuracy']
ax = plt.figure(figsize = (10,6))
p = ax.add_subplot(211)
p.plot(loss)
p = ax.add_subplot(212)
p.plot(acc, color = 'r')
df = pd.DataFrame(hist.history)
df.tail()
```

```python
test_loss , test_acc = model_1.evaluate(X,Y, verbose = 2)
print('\nTest accuracy:', test_acc) 
```

```python
#3 class linear classification
from sklearn.datasets import make_blobs
from pandas import DataFrame
X, y = make_blobs(n_samples=1000, centers=3, n_features=2)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'orange', 1:'pink', 2:'yellow'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
#Activation function=relu,sigmoid
#Loss function=MSE
#Optimizer=Adam
model_1 = keras.Sequential([keras.layers.Dense(32,activation='relu',input_shape = (3,))])
model_1 = keras.Sequential([keras.layers.Dense(32,activation='relu')])
model_1 = keras.Sequential([keras.layers.Dense(16,activation='relu')])
model_1 = keras.Sequential([keras.layers.Dense(1,activation='sigmoid')])
model_1.compile(optimizer = 'Adam', loss = tf.keras.losses.MeanSquaredError(), metrics = ['accuracy'])
hist = model_1.fit(X,y, epochs = 1000, verbose = 2)
```

```python
loss = hist.history['loss']
acc = hist.history['accuracy']
ax = plt.figure(figsize = (10,6))
p = ax.add_subplot(211)
p.plot(loss)
p = ax.add_subplot(212)
p.plot(acc, color = 'r')
df = pd.DataFrame(hist.history)
df.tail()
```

```python
test_loss ,test_acc = model_1.evaluate(X,y, verbose = 2)
print('\nTest accuracy:', test_acc)
```
